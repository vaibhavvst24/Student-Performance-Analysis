```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("student_habits_performance.csv")
data.head()
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
print(df.corr(numeric_only=True)['exam_score'].sort_values(ascending=False))
```

```python
df.fillna({'parental_education_level': df['parental_education_level'].mode()[0],
           'study_hours_per_day': df['study_hours_per_day'].median()}, inplace=True)
```

# Feature Engineering

```python
df['Total_Screen_time'] = df['social_media_hours'] + df['netflix_hours']
```

```python
df['performance'] = pd.cut(df['exam_score'], 
                                bins=[0, 50, 70, 85, 100], 
                                labels=['Low', 'Medium', 'Good', 'Excellent'])
```

```python
df['study_bin'] = pd.cut(df['study_hours_per_day'], 
                         bins=[0, 2, 4, 6, 10], 
                         labels=['0–2', '2–4', '4–6', '6+'])
```

```python
df['sleep_category'] = pd.cut(df['sleep_hours'], 
                              bins=[0, 5, 7, 9, 12], 
                              labels=['<5 hrs', '5–7 hrs', '7–9 hrs', '9+ hrs'])
```

# EDA Exploratory Data Analysis

## Do students with attendance < 75% score lower?

```python
low_attendance = df[df['attendance_percentage'] < 75]['attendance_percentage'].mean()
high_attendance = df[df['attendance_percentage'] >= 75]['attendance_percentage'].mean()

print(f"Low Attendance (<75%): {low_attendance:.2f}")
print(f"High Attendance (>=75%): {high_attendance:.2f}")
```

## Is there a visible pattern between sleep hours and final grades?

```python
plt.figure(figsize = (13,5), dpi = 150, facecolor='k', edgecolor='k')
sns.scatterplot(data=df,x='sleep_hours', y='exam_score', hue='sleep_hours', palette='cubehelix')
plt.xlabel('Sleep Hours per day', fontsize = 12, color='w')
plt.ylabel('Exam Score', fontsize = 12, color='w')
plt.title("Sleep Hours vs Final result", fontsize = 12, color='w')
plt.tick_params(color='w', labelcolor='w')
plt.show()
```

## Do students with better health ratings score better on average?

```python
avg_scores = df.groupby('mental_health_rating')['exam_score'].mean().sort_values(ascending=False)
avg_scores
```

```python
# Bar plot
plt.figure(figsize=(13,5), dpi=140)
sns.barplot(x=avg_scores.index, y=avg_scores.values, palette='rocket', legend=False, hue=avg_scores.index)
plt.title("Average Final Grade by Health Rating")
plt.xlabel("Health Rating")
plt.ylabel("Average Final Grade")
plt.show()
```

## What is the average sleep duration among top-performing students?

```python
# Option A: Based on GradeCategory
top_students = df[df['extracurricular_participation'] == 'Yes']
print("Average sleep hours (GradeCategory = Yes):", top_students['sleep_hours'].mean())

# Option B: Based on FinalGrade
top_students = df[df['exam_score'] >= 90]
print("Average sleep hours (FinalGrade >= 90):", top_students['sleep_hours'].mean())
```

## How does study time affect final grades?

```python
plt.figure(figsize=(13,5), dpi=140, facecolor='k')
sns.regplot(data=df, x='study_hours_per_day', y='exam_score', scatter_kws={'alpha':0.9}, line_kws={'color':'darkslategray'})
plt.title("Study Hours vs Final Grade", fontsize = 14, color='w')
plt.xlabel("Study Hours per Day", fontsize = 12, color='w')
plt.ylabel("Final Grade", fontsize = 12, color='w')
plt.tick_params(color='w', labelcolor='w')
plt.show()
```

## Is there an optimal range of study hours for high performance?

```python
# Create bins for study hours
df['StudyHourGroup'] = pd.cut(df['study_hours_per_day'], bins=[0, 2, 4, 6, 8, 10], labels=['0–2', '2–4', '4–6', '6–8', '8–10'])

# Boxplot
plt.figure(figsize=(13,5), dpi=150, facecolor='k')
sns.boxplot(data=df, x='StudyHourGroup', y='exam_score', palette='dark', hue='StudyHourGroup')
plt.title("Final Grade by Study Hour Group", fontsize = 14, color='w')
plt.xlabel("Study Hours (per day)", fontsize = 12, color='w')
plt.ylabel("Final Grade", fontsize= 12, color='w')
plt.tick_params(color='w', labelcolor='w')
plt.show()
```

## What is the relationship between study hours per day and exam scores?

```python
correlation = df['study_hours_per_day'].corr(df['exam_score'])
print("Correlation between study hours and exam score:", correlation)
```

```python
plt.figure(figsize=(13,5), dpi=140, facecolor='k')
sns.scatterplot(x='study_hours_per_day', y='exam_score', data=df, hue='study_hours_per_day', palette='dark', legend=False)
plt.title('Study Hours vs Exam Score', fontsize = 14, color = 'w')
plt.xlabel("Study Hours", fontsize = 12, color='w')
plt.ylabel("Exam score", fontsize = 12, color='w')
plt.tick_params(color='w', labelcolor='w')
plt.show()
```

## Is there an optimal range of study hours that maximizes exam scores?

```python
plt.figure(figsize=(13,5), dpi=150, facecolor = 'k')
sns.barplot(data=df, x='study_bin', y='exam_score', hue='study_bin', palette='mako')
plt.title('Average Exam Score by Study Hour Ranges', fontsize=14, color='w')
plt.xlabel("Study bin", fontsize = 12, color='w')
plt.ylabel("Exam score", fontsize = 12, color='w')
plt.tick_params(color='w', labelcolor='w')
plt.show()
```

## How many top scorers study less than 3 hours a day?

```python
top_scores = df[df['exam_score'] >= df['exam_score'].quantile(0.9)]
low_study = top_scores[top_scores['study_hours_per_day'] < 3]
print("Number of top scorers who study < 3 hours:", len(low_study))
```

## Is exam performance affected more by attendance or study hours?

```python
corr_study = df['study_hours_per_day'].corr(df['exam_score'])
corr_attendance = df['study_hours_per_day'].corr(df['exam_score'])

print(f"Study Hours vs Exam Score Correlation: {corr_study}")
print(f"Attendance vs Exam Score Correlation: {corr_attendance}")
```

##  Do students with low attendance but high study hours still perform well?

```python
low_attendance = df['attendance_percentage'] < 60
high_study = df['study_hours_per_day'] >= 5

subset = df[low_attendance & high_study]
print("Average score of low attendance & high study students:", subset['exam_score'].mean())
```
